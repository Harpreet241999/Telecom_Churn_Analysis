```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df.info()
```

#Replacing blanks with 0 as tenure is 0 and no chargers are recorded

```python
df['TotalCharges']=df['TotalCharges'].replace(' ','0')
df['TotalCharges']=df['TotalCharges'].astype('float')
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value == 1:
     return "yes"
    else:
     return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

#converted 0 and 1 value to yes/no for easy understanding

```python
df.head()
```

```python
ax = sns.countplot(x = "Churn", data = df)
ax.bar_label(ax.containers[0])
plt.title('Count of Customer')
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb = df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'], labels =gb.index, autopct='%1.2f%%')
plt.title('Percentage of Churned Customer', fontsize =8)
plt.show()
```

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df, hue='Churn')
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x="tenure", data = df, bins=50, hue='Churn')
plt.show()
```

#people who have used the service for long time have stayed and people who used it for short period for 1 or 2 months have churned


```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "Contract", data = df, hue='Churn')
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Contract")
plt.show()
```

#people who have month to month contract are likely to churn then from those who have 1 or 2 month

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x ='PaymentMethod', data = df, hue='Churn')
ax.bar_label(ax.containers[0])
plt.title('Churned Customer Payment Method')
plt.xticks(rotation=45)
plt.show()
```

#Customer is more likely to churn when using electronic check as a payment method
